## 10 Key EDA Findings

| # | Insight | Chart |
|:-:|---------|:-----:|
| **1** | **SIP nearly tripled in 4 years:** Monthly SIP inflows surged from Rs.11,423 Cr (Jan 2022) to an all-time high of **Rs.31,002 Cr (Dec 2025)** — a 171% absolute rise driven by SIP app adoption and rising financial literacy in Tier-2/3 cities. | Chart 3 |
| **2** | **SBI MF's unmatched AUM dominance:** SBI MF reached **Rs.12.5 Lakh Crore** in 2025 — nearly 2x HDFC MF (Rs.7L Cr) — holding ~20% market share on the back of its branch network and sovereign-brand trust. | Chart 2 |
| **3** | **2023 Bull Run was broad-based:** All 10 fund categories posted strong NAV appreciation in 2023, with Small Cap and Mid Cap estimated at **35-45% annual returns** — the strongest calendar year performance in the 4-year window. | Charts 1, 11 |
| **4** | **2024 H1 correction was risk-selective:** The mid-2024 drawdown hit Small Cap / Mid Cap hardest (est. 8-15% NAV decline) while Large Cap and Debt funds held steady — a classic risk-off rotation that rewarded asset allocation discipline. | Charts 1, 14 |
| **5** | **Folio count doubled — a democratisation milestone:** Total folios grew **13.26 Cr → 26.12 Cr** (97% in 4 years), signalling mass-market adoption with new-to-MF investors from Tier-2/3 cities entering the market at scale. | Charts 8, 15a |
| **6** | **Financial Services anchors every equity portfolio:** BFSI commands **~28.5%** of aggregate equity fund weight — reflecting India's ongoing credit-cycle expansion and the financialisation of household savings. | Chart 10 |
| **7** | **SIP inflows are geographically concentrated:** Maharashtra + Delhi NCR + Gujarat alone account for **~41%** of total SIP value; T30 cities contribute 72% — B30 penetration remains the single largest untapped growth lever for AMCs. | Chart 7 |
| **8** | **The 26-45 cohort drives 60% of SIP volumes:** Working-age segments (26-35 and 36-45) dominate SIP registrations, with median ticket sizes rising systematically with age — a healthy lifecycle-investing pattern emerging in retail. | Charts 5 & 6 |
| **9** | **Diversification benefits collapse in stressed markets:** Cross-category return correlations averaged **0.65+** during the 2024 correction, confirming that multi-fund portfolios provide the least shelter precisely when investors need it most. | Chart 9 |
| **10** | **Small Cap volatility is 4-5x that of Debt:** Annualised rolling vol for Small Cap averages **22-28%** versus 4-6% for Debt funds — yet SIP registrations into Small Cap kept rising, exposing a significant behavioural risk-appetite mismatch in retail investors. | Chart 14 |
